For this problem i need to break down a larger number by the prime factors. So 100 would break down this way

2*50
 /  \
2   25
   / \
  5  5
  
So the final factorized version would be 2, 2, 5, 5

```python
def getAllPrimeFactors(n):
    
    # start by setting up the list where the solutions would be stored
    prime_factors = []
    
    # storing the value of the second number in case it needed to be factorized again
    large_number = 0
    
    # these 4 if statements cover the more fringe cases
    if isinstance(n, int) is False:
        return []
    if n == 0:
        return []
    if n == 1:
        return [1]
    if n == 2:
        return[2]
    
    # This loop will start at 2, the smallest prime number available (1 is not included)
    # and it will store the smaller value in the list and the larger value seperately
    for x in range(2, int((n // 2 +1))):
        if n % x == 0:
            y = int(n/x)
            prime_factors.append(x)
            large_number = y
            break
        
        elif x >= int((n // 2)):
            prime_factors.append(n)
            return prime_factors
               
            
    return prime_factors + getAllPrimeFactors(large_number)

# the loop above uses recursion, so the larger number that is stored is put through the 
# loop again until it hits the elif statement and breaks out

        

# this function is to find the count of each value to organize it more cleanly
# ex: 100 -> 2, 2, 5, 5 -> [[2, 5],[2, 2]]
def getUniquePrimeFactorsWithCount(n):
    final_list = []
    values = []
    counts = []
    
    if n == 0:
        return [[],[]]
    
    # simple for loop that uses the list solution obtained in the previous function
    # just used that to get the count of each unique value
    for x in set(getAllPrimeFactors(n)):
        values.append(x)
        counts.append(getAllPrimeFactors(n).count(x))
    final_list.append(values)
    final_list.append(counts)
    
    return final_list
        
    

# this final function is set up to get the products of the prime factors and their count
# ex: 100 [[2, 5],[2, 2]] -> [4, 25]
def getUniquePrimeFactorsWithProducts(n):
    factor_products = []
    
    if n == 0:
        return []
    
    new_n = getUniquePrimeFactorsWithCount(n)
    
    # simple loop for each interior list 
    for i in range(0, len(new_n[0])):
        factor_products.append(new_n[0][i] ** new_n[1][i])
        
    return factor_products
```
